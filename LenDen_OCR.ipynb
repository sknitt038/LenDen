```python
import matplotlib.pyplot as plt
```

```python
import keras_ocr
```

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
images = [
    keras_ocr.tools.read(url) for url in [
       '/home/suresh/Desktop/ocr_/PM.jpeg',
        '/home/suresh/Desktop/ocr_/PAN-Card.jpeg',
    ]
]
```

```python
prediction_groups = pipeline.recognize(images)
```

```python
fig, axs = plt.subplots(nrows=len(images), figsize=(30, 30))
for ax, image, predictions in zip(axs, images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```

```python
info=[]
no_of_data= len(prediction_groups)
for i in range(0, no_of_data):
    for text in prediction_groups[i]:
        info.append(text[0])
```
